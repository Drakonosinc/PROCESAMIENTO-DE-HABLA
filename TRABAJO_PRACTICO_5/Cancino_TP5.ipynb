### cargo las librerias que utilizaré

```python
# NumPy para operaciones numéricas eficientes y manejo de arreglos
import numpy as np
# pandas para manipulación y análisis de datos mediante DataFrames
import pandas as pd
# Importo el módulo 're' para trabajar con expresiones regulares como búsqueda, reemplazo y limpieza de texto
import re
# Importo unicodedata para normalizar cadenas Unicode y facilitar la remoción de acentos
import unicodedata
# spaCy para procesamiento de lenguaje natural y uso de modelos lingüísticos
import spacy
# TfidfVectorizer para transformar colecciones de documentos en matrices TF-IDF.
from sklearn.feature_extraction.text import TfidfVectorizer
# cosine_similarity para calcular la similitud coseno entre vectores
from sklearn.metrics.pairwise import cosine_similarity
```

## Crear el chatbot utilizando TFIDF y similitud del coseno

```python
# Crear DataFrame
df = pd.DataFrame(preguntas_respuestas)
df.head()
```

```python
# Defino una función para limpiar y normalizar texto (quita nulos, mayúsculas, acentos, puntuación y espacios extra)
def limpiar_texto(texto):
    # Si el valor es NaN lo convierto a cadena vacía para evitar errores posteriores
    if pd.isna(texto): return ""
    # Convierto todo a minúsculas para homogeneizar comparaciones y búsquedas
    texto = texto.lower()
    # quito los acentos:
    # - Normalizo la cadena a forma NFD (descompone caracteres compuestos en base + marcas)
    # - Elimino los caracteres cuya categoría Unicode es 'Mn' (marca, como tilde o acento)
    texto = ''.join(c for c in unicodedata.normalize('NFD', texto) if unicodedata.category(c) != 'Mn')
    # elimino cualquier carácter que no sea una letra a-z, un dígito o un espacio
    # Esto quita puntuación, signos de interrogación/exclamación invertidos, símbolos, etc
    texto = re.sub(r'[^a-z0-9\s]', '', texto)
    # Remuevo espacios repetidos y espacios al inicio/fin dejando una sola separación entre palabras
    texto = ' '.join(texto.split())
    # Devuelvo el texto limpio
    return texto

# Aplico la función de limpieza a la columna 'preguntas' del DataFrame
df['preguntas'] = df['preguntas'].apply(limpiar_texto)

# Muestro las primeras filas del DataFrame
df.head()
```

```python
# inicializo el vectorizador TF-IDF
vectorizer = TfidfVectorizer()

# ajusto el vectorizador con todas las preguntas del DataFrame y transformo ese texto en una matriz TF-IDF
tfidf_matrix = vectorizer.fit_transform(df['preguntas'])
```

```python
# defino la función del chatbot que recibe una consulta y devuelve la respuesta más similar junto con la similitud
def chatbot_tfidf(consulta):
    # limpio y normalizo la consulta con la función de preprocesamiento definida anteriormente
    consulta_limpia = limpiar_texto(consulta)
    # transformo la consulta limpia a un vector TF-IDF usando el vectorizador ya ajustado
    consulta_vec = vectorizer.transform([consulta_limpia])
    # calculo la similitud coseno entre la consulta y todas las preguntas del corpus
    similitudes = cosine_similarity(consulta_vec, tfidf_matrix).flatten()
    # extraigo la similitud máxima para evaluar si existe una coincidencia suficientemente buena
    max_sim = max(similitudes)
    # si la similitud máxima está por debajo del umbral, devuelvo un mensaje de no comprensión y la puntuación
    if max_sim < 0.5: return "Lo siento, no entiendo tu pregunta. ¿Puedes reformularla?", max_sim
    # obtengo el índice de la pregunta con mayor similitud
    indice = np.argmax(similitudes)
    # devuelvo la respuesta asociada a ese índice y la similitud máxima
    return df["respuestas"][indice], max_sim
```

```python
# hago una prueba para ver que tal funcina
respuesta_tfidf, similitud_tfidf = chatbot_tfidf("hola")
print(f"Chatbot TF-IDF: {respuesta_tfidf} (Similitud: {similitud_tfidf:.4f})")
```

### Crear otro chatbot utilizando embeddings

elegí spaCy porque es ligero y soporta español

```python
# cargo el modelo de spaCy en español.
nlp = spacy.load("es_core_news_md")
```

```python
# genero los embeddings para cada pregunta del DataFrame usando el modelo de spaCy.
# recorro cada pregunta ya limpiada y obtengo su vector semántico
embeddings_preguntas = [nlp(p).vector for p in df["preguntas"]]
```

```python
# defino la función del chatbot basado en embeddings, que calcula similitud semántica entre la consulta y las preguntas
def chatbot_embeddings(consulta):
    # limpio la consulta para asegurar consistencia con el preprocesamiento aplicado a las preguntas
    consulta_limpia = limpiar_texto(consulta)
    # obtengo el embedding (vector semántico) de la consulta usando el modelo spaCy
    consulta_emb = nlp(consulta_limpia).vector
    # calculo la similitud coseno entre el vector de la consulta y cada embedding de las preguntas almacenadas
    # para cada embedding, genero una matriz 1x1 con el valor de similitud
    similitudes = [cosine_similarity([consulta_emb], [emb])[0][0] for emb in embeddings_preguntas]
    # obtengo la mayor similitud para evaluar si alguna pregunta coincide suficientemente bien
    max_sim = max(similitudes)
    # si la similitud máxima es menor al umbral definido (0.5), devuelvo un mensaje indicando que no hay coincidencias claras
    if max_sim < 0.5: return "Lo siento, no entiendo tu pregunta. ¿Puedes reformularla?", max_sim
    # identifico el índice de la pregunta con mayor similitud
    indice = np.argmax(similitudes)
    # devuelvo la respuesta correspondiente a esa pregunta junto con la similitud máxima
    return df["respuestas"][indice], max_sim
```

```python
# pruebo tambien para ver su funcionamiento
respuesta_embedding, similitud_embedding = chatbot_embeddings("hola como estas")
print(f"Chatbot embedding: {respuesta_embedding} (Similitud: {similitud_embedding:.4f})")
```

### Muestra ambos chatbots funcionando

```python
# defino las preguntas para testear los modelos
preguntas_test = [
    "¿Cuáles son los requisitos para la admisión?",
    "¿Cómo inscribirme en la uni?",
    "¿Becas para alumnos?",
    "¿Horarios de exámenes?",
    "¿Qué es el cambio climático?",
    "¿Cómo cambio contraseña?",
    "¿Programas de intercambio?",
    "¿Normas de la universidad?",
    "¿Dónde está la biblioteca?",
    "¿Fechas para inscribir materias?"]

# defino una función para comparar las respuestas de los dos chatbots (TF-IDF y Embeddings) sobre un conjunto de preguntas
def comparar_bots_df(preguntas_usuario):
    # inicializo una lista donde guardaré los resultados generados por cada pregunta evaluada
    resultados = []
    # recorro cada pregunta de la lista
    for pregunta in preguntas_usuario:
        # limpio la pregunta para mantener coherencia con el preprocesamiento de ambos chatbots
        pregunta_limpia = limpiar_texto(pregunta)
        # obtengo la respuesta del chatbot basado en TF-IDF
        resp_tfidf = chatbot_tfidf(pregunta_limpia)
        # obtengo la respuesta del chatbot basado en embeddings semánticos
        resp_emb = chatbot_embeddings(pregunta_limpia)
        # agrego los resultados en forma de diccionario para construir el DataFrame
        resultados.append({
            "Pregunta": pregunta,
            "Respuesta TF-IDF": resp_tfidf,
            "Respuesta Embeddings": resp_emb})

    # convierto la lista de resultados en un DataFrame para visualizar y analizar comparativamente ambas respuestas
    df_resultados = pd.DataFrame(resultados)
    # configuro pandas para mostrar el contenido completo de las columnas sin recortes
    pd.set_option('display.max_colwidth', None)
    # devuelvo el DataFrame con el resumen comparativo
    return df_resultados
```

```python
# ejecuto la función de comparación usando el conjunto de preguntas de prueba y guardo el resultado en un DataFrame
df_comparacion = comparar_bots_df(preguntas_test)

# muestro el DataFrame resultante con las respuestas generadas por ambos chatbots
display(df_comparacion)
```

## c) Implementación de la Clase ChatBot (RAG)

Crearemos una clase `RAGChatBot` que encapsule toda la lógica:
1.  **`__init__`**: Carga los modelos (LLM y Embeddings) y construye la base de datos vectorial.
2.  **`_build_vector_db`**: Procesa los documentos (respuestas), genera sus embeddings y los almacena en un índice FAISS.
3.  **`retrieve`**: Dada una consulta, genera su embedding y busca los `k` documentos más similares en FAISS.
4.  **`generate_answer`**: Construye el *prompt* (usando la consulta y el contexto recuperado) y se lo pasa al LLM para generar la respuesta final.
5.  **`ask`**: Es el método público que orquesta todo el proceso (retrieve -> generate).

```python
class RAGChatBot:
    def __init__(self, llm_name, embedding_name, documents):
        print(f"Iniciando RAGChatBot con LLM: {llm_name} y Embeddings: {embedding_name}")

        # 1. Cargar Modelo de Embeddings
        print("Cargando Modelo de Embeddings...")
        self.embedding_model = SentenceTransformer(embedding_name)
        print("Modelo de Embeddings cargado.")

        # 2. Cargar LLM
        print("Cargando LLM...")
        print("  - Cargando modelo generativo...")

        self.generator = pipeline("text-generation",
                                  model=llm_name,
                                  max_length=256,
                                  device=-1)
        print("LLM cargado.")

        # 3. Limpiar y almacenar documentos
        self.documents = documents
        self.documentos_limpios = [limpiar_texto(doc) for doc in self.documents]

        # 4. Construir Base de Datos Vectorial (FAISS)
        print("Construyendo Vector DB (FAISS)...")
        self._build_vector_db()
        print("Vector DB construida exitosamente.")

    def _build_vector_db(self):
        # Generar embeddings para todos los documentos de conocimiento
        embeddings = self.embedding_model.encode(self.documentos_limpios,
                                                 normalize_embeddings=True,
                                                 show_progress_bar=True)

        # Obtener la dimensionalidad de los embeddings
        dimension = embeddings.shape[1]

        # Crear un índice FAISS (IndexFlatL2 mide la distancia L2)
        self.index = faiss.IndexFlatL2(dimension)

        # Agregar los embeddings al índice
        self.index.add(embeddings.astype('float32'))

        print("Sistema RAG listo\n")

    def retrieve(self, query, k=3):
        """Recupera los k documentos más relevantes para una consulta."""
        # Limpiar la consulta
        query_limpia = limpiar_texto(query)

        # Generar embedding para la consulta
        query_embedding = self.embedding_model.encode([query_limpia], normalize_embeddings=True)

        # Buscar en el índice FAISS
        # D = distancias, I = índices
        D, I = self.index.search(query_embedding.astype('float32'), k)

        # Retornar documentos recuperados
        retrieved_docs = []
        for idx, dist in zip(I[0], D[0]):
            retrieved_docs.append({'documento': self.documentos_limpios[idx],
                                   'distancia': float(dist)})

        return retrieved_docs

    def generate_answer(self, query, context):
        """Genera una respuesta usando el LLM basado en la consulta y el contexto."""

        # Crear el prompt para el modelo
        prompt = f"""
Contexto:
{context}

Pregunta:
{query}

Por favor, responde la pregunta basándote únicamente en el contexto proporcionado.

Respuesta:
"""

        # Generar la respuesta
        generated = self.generator(prompt, max_new_tokens=100)[0]['generated_text']
        answer = generated.split("Respuesta:")[-1].strip()

        return answer

    def ask(self, query, k=3, model=""):
        """Orquesta el proceso completo de RAG."""
        # 1. Recuperar contexto
        retrieved_docs = self.retrieve(query, k=k)

        print(f"Pregunta: {query}\n")
        print("Documentos recuperados:")
        for i, doc in enumerate(retrieved_docs, 1): print(f"\n  {i}. [Distancia: {doc['distancia']:.4f}]\n\n {doc['documento']}\n")

        # Unir los documentos en un solo string de contexto
        context_str = "\n---\n".join([doc['documento'] for doc in retrieved_docs])

        # 2. Generar respuesta
        answer = self.generate_answer(query, context_str)

        print(f"modelo {model} Respuesta: {answer}\n")
        print("-" * 80 + "\n")

        # Devolvemos la respuesta y los documentos recuperados (para evaluación)
        return answer, retrieved_docs
```

# e) BONUS (opcional)
Context Recall (Exhaustividad): Mide si el sistema fue capaz de encontrar el documento correcto.

Context Precision (Precisión): Penaliza si el documento correcto aparece, pero aparece "abajo" en la lista.

```python
# =============================================================================
# EVALUACIÓN DE MÉTRICAS: CONTEXT PRECISION Y CONTEXT RECALL (MANUAL)
# =============================================================================

def evaluar_retrieval(chatbot, df_eval, nombre_modelo):
    """
    Calcula Context Recall y Context Precision comparando los documentos recuperados
    con el ground_truth (que sabemos que existe en la base de conocimiento).
    """
    resultados = []
    print(f"Iniciando evaluación para: {nombre_modelo}...")

    for index, row in df_eval.iterrows():
        pregunta = row['question']
        ground_truth = row['ground_truth']

        # 1. Obtenemos los documentos recuperados del chatbot (k=3)
        # Nota: Usamos 'ask' pero solo nos importan los docs recuperados, no la respuesta generada por el LLM
        _, retrieved_docs = chatbot.ask(pregunta, k=3)

        # Extraemos solo el texto de los documentos y lo limpiamos para comparar
        textos_recuperados = [limpiar_texto(doc['documento']) for doc in retrieved_docs]
        ground_truth_limpio = limpiar_texto(ground_truth)

        # --- CÁLCULO DE MÉTRICAS ---

        # 2. Context Recall (Sensibilidad)
        # ¿La respuesta correcta está presente en ALGUNO de los documentos recuperados?
        # 1 = Sí, 0 = No.
        is_retrieved = 1 if ground_truth_limpio in textos_recuperados else 0
        context_recall = is_retrieved

        # 3. Context Precision (Precisión del contexto)
        # Evaluamos qué tan "arriba" apareció el documento correcto.
        # Si está 1ro (rank 1) -> 1.0
        # Si está 2do (rank 2) -> 0.5
        # Si está 3ro (rank 3) -> 0.33
        # Si no está -> 0.0
        context_precision = 0.0
        if is_retrieved:
            rank = textos_recuperados.index(ground_truth_limpio) + 1
            context_precision = 1 / rank

        resultados.append({
            'Modelo': nombre_modelo,
            'Pregunta': pregunta,
            'Context Recall': context_recall,
            'Context Precision': round(context_precision, 4)
        })

    return pd.DataFrame(resultados)

# --- EJECUCIÓN DE LA EVALUACIÓN ---

# 1. Evaluar Chatbot 1 (Embeddings: intfloat/multilingual-e5-large)
df_metrics_1 = evaluar_retrieval(chatbot_1, df_evaluacion, "Chatbot 1 (E5-Large)")

# 2. Evaluar Chatbot 2 (Embeddings: BAAI/bge-m3)
df_metrics_2 = evaluar_retrieval(chatbot_2, df_evaluacion, "Chatbot 2 (BGE-M3)")

# 3. Concatenar y mostrar promedio de resultados
df_final_metrics = pd.concat([df_metrics_1, df_metrics_2], ignore_index=True)

# Calcular los promedios por modelo
resumen_metricas = df_final_metrics.groupby('Modelo')[['Context Recall', 'Context Precision']].mean()

print("\n" + "="*50)
print("RESULTADOS FINALES DE EVALUACIÓN DE EMBEDDINGS")
print("="*50)
display(resumen_metricas)

print("\n" + "="*50)
print("DETALLE POR PREGUNTA")
print("="*50)
display(df_final_metrics)
```

## Context Recall
El modelo Chatbot 2 alcanzó un Context Recall perfecto de 1.0, lo que significa que recuperó la información relevante en el 100% de las preguntas mientras que por el contrario, el Chatbot 1 obtuvo un 0.8, fallando en recuperar el contexto en el 20% de los casos.

## Context Precision
El Chatbot 2 también lidera en Context Precision con un 0.816 vs 0.750, lo cual podria indicar que los fragmentos de texto que el Chatbot 2 entrega al LLM son más relevantes y contienen menos ruido o información innecesaria que los entregados por el Chatbot 1.

### Conclusion
Estos datos refuerzan el uso del modelo 2 ya que mientras que el Chatbot 1 falla estructuralmente en encontrar la información en 2 de cada 10 consultas, el Chatbot 2 demuestra ser un RAG mucho más robusto y fiable, garantizando que el LLM siempre tenga acceso a los datos necesarios para responder.

# f) BONUS (opcional)

modelo de embeddings ganador (chatbot_2 con BAAI/bge-m3):

Answer Relevancy (Proxy): Calcularemos la Similitud Coseno entre el vector de la Pregunta y el vector de la Respuesta Generada. Si la respuesta responde lo que se preguntó, deben estar cerca semánticamente.

Faithfulness (Proxy): Calcularemos la Similitud Coseno entre el vector de la Respuesta Generada y el vector del Contexto Recuperado. Una alta similitud indica que la respuesta se adhiere (es fiel) a la información encontrada y no es una alucinación externa.

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def evaluar_generacion_llm(chatbot, df_eval):
    """
    Evalúa la calidad de la generación del LLM usando métricas vectoriales.
    Utiliza el modelo de embeddings del propio chatbot para medir distancias semánticas.
    """
    resultados = []
    print(f"Iniciando evaluación de generación con el modelo de embeddings: {chatbot.embedding_model.modules}")

    for index, row in df_eval.iterrows():
        pregunta = row['question']

        # 1. Obtener respuesta y contexto del chatbot
        # (Usamos k=3 porque fue tu configuración estándar)
        respuesta_generada, docs_recuperados = chatbot.ask(pregunta, k=3, model="Evaluación")

        # Unimos el texto del contexto recuperado para vectorizarlo
        contexto_unido = " ".join([d['documento'] for d in docs_recuperados])

        # 2. Generar Embeddings (Vectorización)
        # Necesitamos vectores para: Pregunta, Respuesta y Contexto
        emb_pregunta = chatbot.embedding_model.encode([pregunta])
        emb_respuesta = chatbot.embedding_model.encode([respuesta_generada])
        emb_contexto = chatbot.embedding_model.encode([contexto_unido])

        # 3. Calcular Métricas (Similitud del Coseno)

        # A) Answer Relevancy: ¿Qué tanto se relaciona la respuesta con la pregunta?
        relevancy_score = cosine_similarity(emb_pregunta, emb_respuesta)[0][0]

        # B) Faithfulness: ¿Qué tanto se relaciona la respuesta con el contexto recuperado?
        # (Si el score es bajo, significa que el modelo generó texto ajeno al contexto)
        faithfulness_score = cosine_similarity(emb_respuesta, emb_contexto)[0][0]

        resultados.append({
            'Pregunta': pregunta,
            'Respuesta Generada': respuesta_generada,
            'Answer Relevancy': round(relevancy_score, 4),
            'Faithfulness': round(faithfulness_score, 4)
        })

    return pd.DataFrame(resultados)

# --- EJECUCIÓN ---

# Usamos chatbot_2 porque determinaste en el punto d) que era el mejor
print("Evaluando LLM del Chatbot 2 (Ganador)...")
df_resultados_llm = evaluar_generacion_llm(chatbot_2, df_evaluacion)

# Calcular promedios
promedio_relevancia = df_resultados_llm['Answer Relevancy'].mean()
promedio_fidelidad = df_resultados_llm['Faithfulness'].mean()

print("\n" + "="*60)
print(f"RESULTADOS FINALES DE GENERACIÓN (LLM + Embeddings BAAI/bge-m3)")
print("="*60)
print(f"Promedio Answer Relevancy: {promedio_relevancia:.4f}")
print(f"Promedio Faithfulness:     {promedio_fidelidad:.4f}")

print("\n" + "="*60)
print("DETALLE POR PREGUNTA")
print("="*60)
# Mostramos solo las columnas clave para facilitar la lectura
display(df_resultados_llm[['Pregunta', 'Answer Relevancy', 'Faithfulness']])
```

## Faithfulness
El promedio de 0.8255 indica que el modelo tiene una fuerte tendencia a adherirse al contexto proporcionado y no inventar información externa.

## Answer Relevancy
El puntaje de relevancia es notablemente más bajo 0.62 aproximadamente, lo cual podria confirmar el evidente problema de la suciedad en las respuestas.

### Conclusion
Estos datos pueden indicar que el chatbot 2 es un sistema honesto y desprolijo, ya que su alta fidelidad garantiza que la información base sea correcta, pero su moderada relevancia puede indicar que requiere una capa de limpieza quizas para filtrar las explicaciones internas y las fugas de instrucciones antes de enviar la respuesta.
